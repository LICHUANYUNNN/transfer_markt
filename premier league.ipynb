```python
import unittest
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd 
browser = webdriver.Chrome("C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe")
```

```python
browser.get('https://www.transfermarkt.co.uk/premier-league/startseite/wettbewerb/GB1')
browser.implicitly_wait(3)
result = browser.find_elements_by_xpath('//*[@id="yw1"]/table/tbody/tr')

league = pd.DataFrame(columns = ['club','link']) 

for x in range(1,len(result)+1): 
        
    club_element = browser.find_element_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[3]/a')
    club = club_element.text
  
    link_element = browser.find_element_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[3]/a').get_attribute('href')
    link = link_element
    
    league.loc[x-1] = [club,link] 
```

```python
league['link']=league['link'].str.replace('startseite','kader')
```

```python
league.to_csv("league.csv")
```
