```python
import unittest
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd 
browser = webdriver.Chrome("C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe")
```

```python
browser.get('https://www.transfermarkt.co.uk/liverpool-fc/kader/verein/31/saison_id/2018/plus/1')
browser.implicitly_wait(3)
result = browser.find_elements_by_xpath('//*[@id="yw1"]/table/tbody/tr') 
player = pd.DataFrame(columns = ['number','name','position','birthday','nation','height','foot','joined','signed from','contract until','market value','change']) 
    
for x in range(1,len(result)+1): 
        
    number_element = browser.find_element_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[1]/div') 
    number = number_element.text
        
    name_element = browser.find_elements_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[2]/table/tbody/tr[1]/td[2]/div[1]')[0]
    name = name_element.text 
        
    position_element = browser.find_elements_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[2]/table/tbody/tr[2]/td')[0]
    position = position_element.text
        
    birth_element = browser.find_elements_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[3]')[0]
    birth = birth_element.text
        
    nation_element = browser.find_elements_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[4]/img')[0].get_attribute('alt')
    nation = nation_element
        
    height_element = browser.find_elements_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[5]')[0]
    height = height_element.text
        
    foot_element = browser.find_elements_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[6]')[0]
    foot = foot_element.text
        
    join_element = browser.find_elements_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[7]')[0]
    join = join_element.text

    sign_element = browser.find_elements_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[8]/a/img')[0].get_attribute('alt')
    sign = sign_element
    
    contract_element = browser.find_elements_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[9]')[0]
    contract = contract_element.text
        
    value_element = browser.find_elements_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[10]')[0]
    value = value_element.text

    change_element = browser.find_elements_by_xpath('//*[@id="yw1"]/table/tbody/tr['+str(x)+']/td[10]/span')[0].get_attribute('title')
    change = change_element

    player.loc[len(player)] = [number,name,position,birth,nation,height,foot,join,sign, contract,value,change] 
```
